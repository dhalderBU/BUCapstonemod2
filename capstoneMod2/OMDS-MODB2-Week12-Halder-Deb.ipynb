# 2. Weekly graph question

Suppose you plot the number visitors to a museum over the seven days of a week; then again the following week.  Can you suggest the advantages and disadvantages of the following two ways of presenting the data?

### Advantage of second over 1st way of representing
- the 2nd diagram is better way to represent as both diagram are on the same scale  but the first one has different y axis and even for x axis the limit is slightly bigger in the first diagram

- In the first diagram looks more cluttered as there are multiple axes and the second one is more clear as there is single pair

- since the second diagram is more clear it is easier to interpret and understand the data and cognitive load is less

- based on the description given if we look at the diagram 2 it gives more context

-  The 2nd approach is easier to implement as its single plot  and will take less real estate on the page


### Problem with both approaches:
- Both the diagram doesn't have the title and also legend is missing so we wont know from diagram what is being represented by which line

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
time_series_0 = np.array([100, 110, 130, 120, 100, 90, 80])
time_series_1 = np.array([70, 90, 100, 100, 120, 110, 90])
fig, axs = plt.subplots(1, 2)
axs[0].plot(time_series_0)
axs[1].plot(time_series_1)
plt.subplots_adjust(wspace = 0)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# X-coordinates
years_x_coordinates = [2008, 2009, 2010, 2011, 2012, 2013, 2014]

# y coordinates
y_line_a = np.array([400, 420, 450, 400, 280, 250, 225])
y_line_b = np.array([320, 380, 400, 300, 270, 240, 215])
y_line_c = np.array([None, None, None, 100, 180, 200, 205])
y_line_d = np.array([None, None, None, 180, 250, 220, 180])
y_line_e = np.array([None, None, None, None, None, 100, 180])

# let me find the index where the value starts
index_a = np.where(y_line_a != None)[0][0]
index_b = np.where(y_line_b != None)[0][0]
index_c = np.where(y_line_c != None)[0][0]
index_d = np.where(y_line_d != None)[0][0]
index_e = np.where(y_line_e != None)[0][0]

# add the line plot and then add the annotation at the start
plt.figure(figsize=(12, 7))
plt.plot(years_x_coordinates[index_a:], y_line_a[index_a:], color='gray')
plt.annotate('A', xy=(years_x_coordinates[index_a], y_line_a[index_a]+8), fontsize=14, fontweight='bold', color='gray')
plt.plot(years_x_coordinates[index_b:], y_line_b[index_b:], color='gray')
plt.annotate('B', xy=(years_x_coordinates[index_b], y_line_b[index_b]+8), fontsize=14, fontweight='bold', color='gray')
plt.plot(years_x_coordinates[index_c:], y_line_c[index_c:], color='gray')
plt.annotate('C', xy=(years_x_coordinates[index_c], y_line_c[index_c]+8), fontsize=14, fontweight='bold', color='gray')
plt.plot(years_x_coordinates[index_d:], y_line_d[index_d:], color='gray')
plt.annotate('D', xy=(years_x_coordinates[index_d], y_line_d[index_d]+8), fontsize=14, fontweight='bold', color='gray')
plt.plot(years_x_coordinates[index_e:], y_line_e[index_e:], color='gray')
plt.annotate('E', xy=(years_x_coordinates[index_e], y_line_e[index_e]+8), fontsize=14, fontweight='bold', color='gray')

#let me add the band for the recommendation range
plt.fill_between(years_x_coordinates, 190, 220, color='lightblue', alpha=0.5, label='Recommended range')

plt.title('Retail Price over Time - fig 8.19 Page 203', fontsize=18, weight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Average price', fontsize=12)

plt.ylim(0, 500)
plt.yticks(np.arange(0, 551, 50))
plt.xlim(2008, 2014)

plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
